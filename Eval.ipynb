```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
RHG5_1 = pd.read_csv("RHG-5/DIS-edges.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
RHG5_1["Algorithm"] = "DIS-edges"
RHG5_2 = pd.read_csv("RHG-5/DIS-edges-s.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
RHG5_2["Algorithm"] = "DIS-edges-s"
RHG5_3 = pd.read_csv("RHG-5/DIS-tuples.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
RHG5_3["Algorithm"] = "DIS-tuples"
RHG5_4 = pd.read_csv("RHG-5/DIS-tuples-s.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
RHG5_4["Algorithm"] = "DIS-tuples-s"
RHG5_5 = pd.read_csv("RHG-5/DIS-global.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
RHG5_5["Algorithm"] = "DIS-global"
RHG5_6 = pd.read_csv("RHG-5/Long-DIS-edges-s.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
RHG5_6["Algorithm"] = "Long-DIS-edges-s"
RHG5 = pd.concat([RHG5_1, RHG5_2, RHG5_3, RHG5_4, RHG5_5, RHG5_6])
RHG5["Fraction of Non-Independent Edges"] = 1. - RHG5["Independent Edges"] / (1000*999/2.)

RHG10_1 = pd.read_csv("RHG-10/DIS-edges.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
RHG10_1["Algorithm"] = "DIS-edges"
RHG10_2 = pd.read_csv("RHG-10/DIS-edges-s.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
RHG10_2["Algorithm"] = "DIS-edges-s"
RHG10_3 = pd.read_csv("RHG-10/DIS-tuples.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
RHG10_3["Algorithm"] = "DIS-tuples"
RHG10_4 = pd.read_csv("RHG-10/DIS-tuples-s.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
RHG10_4["Algorithm"] = "DIS-tuples-s"
RHG10_5 = pd.read_csv("RHG-10/DIS-global.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
RHG10_5["Algorithm"] = "DIS-global"
RHG10 = pd.concat([RHG10_1, RHG10_2, RHG10_3, RHG10_4, RHG10_5])
RHG10["Fraction of Non-Independent Edges"] = 1. - RHG10["Independent Edges"] / (1000*999/2.)

G03_1 = pd.read_csv("Gilbert-p0.3/DIS-edges.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
G03_1["Algorithm"] = "DIS-edges"
G03_2 = pd.read_csv("Gilbert-p0.3/DIS-edges-s.out", sep=",", names=["Type", "Thinning", "Snapshots", "Successful Operations", "Independent Edges", "Non-Independent Edges", "Seed"])
G03_2["Algorithm"] = "DIS-edges-s"
G03 = pd.concat([G03_1, G03_2])
G03["Fraction of Non-Independent Edges"] = 1. - G03["Independent Edges"] / (1000*999/2.)
```

```python
sb.lineplot(data=RHG5, x="Thinning", y="Fraction of Non-Independent Edges", hue="Algorithm")
plt.yscale('log')
plt.title("RHG, n=1000, d=5, gamma=2.5")
```

```python
sb.lineplot(data=RHG10, x="Thinning", y="Fraction of Non-Independent Edges", hue="Algorithm")
plt.yscale('log')
plt.title("RHG, n=1000, d=10, gamma=2.5")
```

```python
sb.lineplot(data=G03, x="Thinning", y="Fraction of Non-Independent Edges", hue="Algorithm")
plt.yscale('log')
plt.title("Gilbert, n=1000, p=0.3")
```
