```python
import pandas
import matplotlib.pyplot as plt

df = pandas.read_csv("output/DATA.dat", sep='\t')
df.loc[df['Gen'] == 'PLDHH'].plot(kind='scatter', x='Thinning', y='Independence-Rate', s=5)
plt.title("Powerlaw graph: $n = 1000, \gamma = -2, mindeg = 5, maxdeg = 1000$")
plt.show()
df.loc[df['Gen'] == 'ERDOSRENYI'].plot(kind='scatter', x='Thinning', y='Independence-Rate', s=5)
plt.title("ErdosRenyi graph: $n = 500, p = 0.5$")
plt.show()
```
